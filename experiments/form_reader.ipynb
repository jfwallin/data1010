# Read Form Responses by Group Code
Paste the published CSV URL from the Sheet linked to your Form.


```python
import pandas as pd
from IPython.display import display

CSV_URL = 'PASTE_PUBLISHED_CSV_URL_HERE'  # ends with export?format=csv
group_code = input('Enter your 5-digit group code: ').strip()

try:
    df = pd.read_csv(CSV_URL)
    df.columns = [c.strip().lower() for c in df.columns]
    if 'group_code' not in df.columns:
        raise ValueError('group_code column not found in CSV')
    sub = df[df['group_code'].astype(str) == group_code]
    if sub.empty:
        print('No entries found for that group code.')
    else:
        if 'timestamp' in sub.columns:
            sub = sub.sort_values(by='timestamp', ascending=False)
        cols = [c for c in sub.columns if c not in ('id','group_code')]
        display(sub[['group_code'] + cols])
except Exception as e:
    print('Could not load responses:', e)
```
