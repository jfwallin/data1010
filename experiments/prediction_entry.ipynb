# Prediction Entry to CSV (Form-ready)
Enter your group code and responses. This saves a CSV row you can upload to a Form/Sheet.


```python
import pandas as pd
from datetime import datetime

group_code = input('Enter 5-digit group code: ').strip()
prediction = input('Prediction (short): ').strip()
prediction_long = input('Prediction (full question answer): ').strip()
result_short = input('Result (short): ').strip()
result_long = input('What actually happened? ').strip()
evaluation_short = input('Evaluation (short): ').strip()
evaluation_long = input('Explain your evaluation: ').strip()
notes_short = input('Notes (short): ').strip()
notes_long = input('Notes (long/optional): ').strip()

row = {
    'id': 'student-entry',
    'group_code': group_code,
    'timestamp': datetime.utcnow().isoformat(),
    'prediction_short': prediction,
    'prediction_long': prediction_long,
    'result_short': result_short,
    'result_long': result_long,
    'evaluation_short': evaluation_short,
    'evaluation_long': evaluation_long,
    'notes_short': notes_short,
    'notes_long': notes_long,
}
df = pd.DataFrame([row])
df.to_csv('prediction_entry.csv', index=False)
print('\nSaved to prediction_entry.csv — upload this to your Form/Sheet or attach to LMS.')
```
